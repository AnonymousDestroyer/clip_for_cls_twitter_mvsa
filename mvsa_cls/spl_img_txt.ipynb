```python
import os
import shutil

# 声明图像文件夹和文本文件夹路径
image_folder_path = "data/MVSA_spl/img"
text_folder_path = "data/MVSA_spl/txt"
image_txt_folder_path = "data/MVSA_Single/data"

# 如果文本文件夹或图像文件夹不存在，则创建它们
if not os.path.exists(image_folder_path):
    os.makedirs(image_folder_path)
if not os.path.exists(text_folder_path):
    os.makedirs(text_folder_path)

# 遍历文件夹中的每个文件
for filename in os.listdir(image_txt_folder_path):
    print(filename)
    # 如果是图像文件，则将其移动到图像文件夹中
    if filename.endswith(".jpg"):
        shutil.copy(os.path.join(image_txt_folder_path, filename), os.path.join(image_folder_path, filename))
    # 如果是文本文件，则将其移动到文本文件夹中
    elif filename.endswith(".txt"):
        shutil.copy(os.path.join(image_txt_folder_path, filename), os.path.join(text_folder_path, filename))

print('ok')
```

```python
label_path = './data/MVSA_spl/label.txt'
label_data = open(label_path,'r').readlines()
label_data
```

```python
dic = {}
for label_line in label_data:
    label_line_list = label_line.split()
    labe_index = label_line_list[0]
    text_label = label_line_list[1].split(',')[0]
    image_label = label_line_list[1].split(',')[1]
    print(labe_index,text_label,image_label)
    dic[labe_index] = {"text":text_label,"image":image_label}
dic
```
